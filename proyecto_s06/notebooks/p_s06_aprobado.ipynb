## 2. Preparación de los datos



### Nombre de las columnas a snake_case

```python
# Remplazo los nombres de columnas a snake_case
new_col_names = []

for old_name in df_videojuegos.columns:
    nombre_minusculas = old_name.lower()
    new_col_names.append(nombre_minusculas)

df_videojuegos.columns = new_col_names

df_videojuegos.columns
```

### Transformacion de los datos

```python
# Convierto el año de lanzamiento a datatime
df_videojuegos['year_of_release'] = pd.to_numeric(df_videojuegos['year_of_release'], errors='coerce').astype('Int64')

# Creo una columna adicional que indique si era 'tbd'
df_videojuegos['user_score_tbd'] = df_videojuegos['user_score'] == 'tbd'
# Remplazo tbd a nan
df_videojuegos['user_score'] = df_videojuegos['user_score'].replace('tbd', np.nan)
# Convierto  la puntuacion del usuario a flotante
df_videojuegos['user_score'] = pd.to_numeric(df_videojuegos['user_score'], errors='coerce')
```

Transformación del año de lanzamiento (float → int)

* Se convirtió la columna de año de lanzamiento de flotante a entero porque los años son valores discretos y no requieren decimales.Esta modificación mejora la coherencia en análisis temporales, facilitando operaciones como filtrado o agrupamiento.
* Además, optimiza el uso de memoria al almacenar los datos en un formato más eficiente (int ocupa menos espacio que float).

Transformación de la puntuación de usuario (object → float)

* La columna de puntuación de usuario contenía valores mixtos, incluyendo la etiqueta "tbd", lo que impedía su tratamiento como dato numérico.

    Para solucionarlo:

    * Se creó una columna auxiliar para identificar registros con "tbd".

    * Estos valores se reemplazaron por NaN para evitar sesgos, permitiendo la conversión a float.

* Este proceso asegura que la columna pueda usarse en cálculos estadísticos sin distorsiones por textos no numéricos.

### Valores nulos


```python
# Funcion que analiza y muestra la cantidad con porcentaje de valores ausentes por columna.
def analizar_valores_ausentes(df, nombre_df='el_DataFrame'):
    # Parámetros: ( df= DataFrame a analizar, nombre_df= Nombre del DataFrame )

    # Calculo el total de filas
    total_filas = df.shape[0]

    # Obtengo el conteo de valores ausentes por columna
    valores_ausentes = df.isna().sum()

    print(f"Análisis de valores ausentes en {nombre_df}:")
    print("")

    # Recorrer cada columna y mostrar resultados
    for columna, na_count in valores_ausentes.items():
        if na_count > 0:  # Solo mostrar columnas con valores ausentes
            porcentaje = (na_count / total_filas) * 100
            print(f"Identifiqué {na_count} valores ausentes en la columna '{columna}' "
                  f"({porcentaje:.2f}% del total).")

    print(f"\nColumnas con valores ausentes: {sum(valores_ausentes > 0)} de {len(df.columns)}")

# Uso de la función:
analizar_valores_ausentes(df_videojuegos, 'videojuegos')
```

```python
# Elimino los ausentes en nombre, genero y año de lanzamiento
df_videojuegos.dropna(subset=['name', 'genre','year_of_release'], inplace=True)

# Calculo la mediana de la puntuacion de la critica
mediana_critica = df_videojuegos['critic_score'].median()
df_videojuegos['critic_score'] = df_videojuegos['critic_score'].fillna(mediana_critica)

# Calculo la mediana de la puntuacion del usuario
mediana_usuario = df_videojuegos['user_score'].median()
df_videojuegos['user_score'] = df_videojuegos['user_score'].fillna(mediana_usuario)

# Relleno valores ausentes como desconocido
df_videojuegos['rating'] = df_videojuegos['rating'].fillna('Unk')

analizar_valores_ausentes(df_videojuegos, 'videojuegos')
```

* Para columnas con pocos valores faltantes (nombre 0.01%, género 0.01% y año de lanzamiento 1.6%), opte por eliminar estos registros dado su mínimo impacto en el análisis general. Esta decisión permite mantener la integridad del dataset sin afectar significativamente los resultados.

* En la columna de clasificación ESRB, reemplaze los valores ausentes  con "Unk" (desconocido), considerando que muchos juegos antiguos (previos a 1994) carecían de esta clasificación. Este enfoque preservó los registros históricos sin asignar categorías incorrectas.

* Para las puntuaciones de críticos y usuarios, utilice la mediana para imputar los valores faltantes, ya que esta medida es más robusta ante outliers. A diferencia del promedio, la mediana no se ve distorsionada por calificaciones extremas, lo que garantiza una estimación más representativa del rendimiento típico de los juegos.

### Calculo de las ventas totales de todas las regiones

```python
# Crear columna de ventas globales
df_videojuegos['total_sales'] = df_videojuegos[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
df_videojuegos = df_videojuegos[['name', 'platform', 'year_of_release',	'genre',	'na_sales', 'eu_sales',	'jp_sales',	'other_sales', 'total_sales', 'critic_score',	'user_score', 'rating'	]]

# Agrupo por año de lanzamiento, sumando las ventas totales
ventas_al_año = df_videojuegos.groupby(['year_of_release']).agg(year_sales=('total_sales','sum')).sort_values(by='year_of_release', ascending=True)
ventas_al_año.plot(kind='bar', figsize=(12,6))
plt.title('Ventas al Año')
plt.xlabel('Año')
plt.ylabel('Millones de dolares USD')
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

#####  ¿Cómo varían las ventas de una plataforma a otra? ¿Cuánto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer?

```python
# Agrupo por plataforma y calculo el año promedio y el total de las ventas
ventas_plataforma =  df_videojuegos.groupby(['platform']).agg(median_year=('year_of_release', 'median'),
                                                              total_sales=('total_sales','sum'),
                                                              total_games=('name','count')).sort_values(by='total_sales', ascending=False)

# Selecciono las 10 con mayores ventas
top_ventas_plataforma = ventas_plataforma.head(10).index.tolist()

# Filtro para las plataformas mas vendidas
top_plataforma = df_videojuegos[df_videojuegos['platform'].isin(top_ventas_plataforma)]

# Creo una tabla pivot con años como índice y plataformas como columnas
ventas_anuales = top_plataforma.pivot_table(
    index='year_of_release',
    columns='platform',
    values='total_sales',
    aggfunc='sum'
)

# Visualización
ventas_anuales.plot(kind='line', stacked=True, figsize=(12, 6))
plt.title('Ventas anuales por plataforma')
plt.xlabel('Años')
plt.ylabel('Ventas globales (millones)')
plt.grid(linestyle="--", alpha=0.7)
plt.show()
```

```python
# Obtener el último año con ventas para cada plataforma
ultimo_año_ventas = df_videojuegos.groupby('platform')['year_of_release'].max()

# Obtener el año más reciente en el dataset
año_mas_reciente = df_videojuegos['year_of_release'].max()

# Plataformas sin ventas recientes en 2 años
plataformas_obsoletas = ultimo_año_ventas[ultimo_año_ventas < (año_mas_reciente - 2)].index.tolist()

# Ver su popularidad histórica
ventas_historicas = df_videojuegos[df_videojuegos['platform'].isin(plataformas_obsoletas)].groupby('platform')['total_sales'].sum().sort_values(ascending=False)

# Visualización
ventas_historicas.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Ventas plataformas obsoletas')
plt.xlabel('Años')
plt.ylabel('Millones de dolares USD')
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

```python
# Calculo años de actividad por plataforma
vida_util = df_videojuegos.groupby('platform')['year_of_release'].agg(['min', 'max'])
vida_util['duracion'] = vida_util['max'] - vida_util['min'] + 1

# Estadísticas generales
duracion_promedio = vida_util['duracion'].mean()
duracion_mediana = vida_util['duracion'].median()

# Tiempo entre lanzamiento de nuevas plataformas
años_lanzamiento = df_videojuegos.groupby('platform')['year_of_release'].min().sort_values()
tiempos_entre_lanzamientos = años_lanzamiento.diff().dropna().median()

duracion_mediana, duracion_promedio, tiempos_entre_lanzamientos

print(f'El tiempo de vida util de una plataforma es de {duracion_mediana} años')
print(f'El tiempo entre plataformas es de {tiempos_entre_lanzamientos} años')
```

## 4. Prueba las hipótesis estadísticas

Hipótesis nula (H₀): μ_XboxOne = μ_PC (las calificaciones promedio son iguales)

Hipótesis alternativa (H₁): μ_XboxOne ≠ μ_PC (las calificaciones promedio son diferentes)

```python
# Filtrar datos
xbox_scores = df_periodo_videojuegos[df_periodo_videojuegos['platform'] == 'XOne']['user_score'].dropna()
pc_scores = df_periodo_videojuegos[df_periodo_videojuegos['platform'] == 'PC']['user_score'].dropna()

# Verifico la igualdad de varianzas con Levene
levene_stat, levene_p = stats.levene(
    xbox_scores,
    pc_scores)

alpha = 0.05
equal_var = levene_p > alpha  # True si varianzas iguales, False si diferentes

print(f"Prueba de Levene: p-value = {levene_p:.4f}")
print(f"¿Asumir varianzas iguales? {'Sí' if equal_var else 'No'}")

# Realizar prueba t independiente
t_stat, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=equal_var) 

print(f"Xbox One (M={xbox_scores.mean():.2f}) vs PC (M={pc_scores.mean():.2f})")
print(f"Estadístico t: {t_stat:.3f}, Valor-p: {p_value:.5f}")

if p_value < alpha:
    print("Conclusión: Rechazamos H₀. Hay una diferencia significativa.")
    # Comparar medias para dirección de la diferencia
    if xbox_scores.mean() > pc_scores.mean():
        print("xBox One tiene un ingreso promedio MAYOR que PC.")
    else:
        print("xBox One tiene un ingreso promedio MENOR que PC.")
else:
    print("Conclusión: No podemos rechazar H₀. No hay evidencia de diferencia. Las varianzas no son iguales")
```

Tras realizar la prueba t de comparación de medias con un nivel de confianza estándar (α=0.05), los resultados indican que:

  * No existe evidencia significativa (p=0.2048 > 0.05) para rechazar la hipótesis nula

  * La diferencia observada entre las calificaciones promedio (Xbox One: 6.78 vs PC: 6.62) es estadísticamente irrelevante

Hipótesis nula (H₀): μ_Acción = μ_Deportes (las calificaciones promedio son iguales)

Hipótesis alternativa (H₁): μ_Acción ≠ μ_Deportes (las calificaciones promedio son diferentes)

```python
# Filtrar datos
action_scores = df_periodo_videojuegos[df_periodo_videojuegos['genre'] == 'Action']['user_score'].dropna()
sports_scores = df_periodo_videojuegos[df_periodo_videojuegos['genre'] == 'Sports']['user_score'].dropna()

# Verifico la igualdad de varianzas con Levene
levene_stat, levene_p = stats.levene(
    action_scores,
    sports_scores)

alpha = 0.05
equal_var = levene_p > alpha  # True si varianzas iguales, False si diferentes

print(f"Prueba de Levene: p-value = {levene_p:.4f}")
print(f"¿Asumir varianzas iguales? {'Sí' if equal_var else 'No'}")

# Realizar prueba t independiente
t_stat, p_value = stats.ttest_ind(action_scores, sports_scores, equal_var=equal_var)

print(f"Acción (M={action_scores.mean():.2f}) vs Deportes (M={sports_scores.mean():.2f})")
print(f"t = {t_stat:.3f}, p = {p_value:.5f}")



if p_value < alpha:
    print("Conclusión: Rechazamos H₀. Hay una diferencia significativa entre los generos.")
    # Comparar medias para dirección de la diferencia
    if action_scores.mean() > sports_scores.mean():
        print("El genero de Accion tiene un ingreso promedio MAYOR que Deportes.")
    else:
        print("El genero de Accion tiene un ingreso promedio MENOR que Deportes.")
else:
    print("Conclusión: No podemos rechazar H₀. No hay evidencia de diferencia.")
```
